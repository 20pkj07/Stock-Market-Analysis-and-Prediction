## Volume of Sales
Volume is the amount of an asset or security that changes hands over some period of time, often over the course of a day. For instance, the stock trading volume would refer to the number of shares of security traded between its daily open and close. Trading volume, and changes to volume over the course of time, are important inputs for technical traders.

```python
plt.figure(figsize=(16,20))
for i in range(5):
    plt.subplot(3,2,i+1)
    plt.plot(companies[i]['Volume'])
    plt.grid()
    plt.xlabel('Year')
    plt.ylabel('Volume')
    plt.title(f" {comp[i]} Sale Volume")
plt.tight_layout()
```

# Moving Average
The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks, or any time period the trader chooses

```python
meta.rolling(7).mean().head(10)
```

```python
for i in range(5):
    companies[i]["MA30"]=companies[i]['Close'].rolling(30).mean()
    companies[i]["MA90"]=companies[i]['Close'].rolling(90).mean()
```

```python
for i in range(5):
    companies[i]["MA"]=companies[i]['Close'].rolling(30).mean()
```

```python
plt.figure(figsize=(16,20))
for j in range(5):
    plt.subplot(3,2,j+1)
    plt.grid(True)
    plt.xlabel('Year')
    plt.ylabel('Moving Average')
    plt.title(f" {comp[j]} Moving Average")
    companies[j]['Close'].plot()
    companies[j]['MA30'].plot(linestyle='--')
    companies[j]['MA90'].plot(linestyle='--')
    plt.legend()
plt.tight_layout()
```

# Volatility Calculation
The volatility of a stock is a measurement of the change in variance in a stock's returns over a specific period.
We can compare the volatilities of stocks to examine which are riskier.

```python
for i in range(5):
    companies[i]['TrueRange'] = companies[i]['High'] - companies[i]['Low']
    companies[i]['ATR'] = companies[i]['TrueRange'].rolling(window=14).mean()
```

```python
plt.figure(figsize=(16,20))
for j in range(5):
    plt.subplot(3,2,j+1)
    companies[j]['ATR'].plot()
    plt.xlabel('Date')
    plt.ylabel('ATR')
    plt.title(f" {comp[j]} Volatility Analysis")
    plt.grid()
plt.tight_layout()
```

```python
plt.figure(figsize=(30,40))
for j in range(5):
    plt.subplot(3,2,j+1)
    correlation_matrix = companies[j].corr()
    sns.heatmap(correlation_matrix, annot=True)
    plt.title(f" {comp[j]} Correlation Heatmap")
plt.tight_layout()
```

# Training

```python
meta_tn=pd.DataFrame(companies[0]['Open'])
amazon_tn=pd.DataFrame(companies[1]['Open'])
apple_tn=pd.DataFrame(companies[2]['Open'])
netflix_tn=pd.DataFrame(companies[3]['Open'])
google_tn=pd.DataFrame(companies[4]['Open'])
```

```python
meta_tn
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler (feature_range= (0,1))

meta_scaled=scaler.fit_transform(meta_tn)
amazon_scaled=scaler.fit_transform(amazon_tn)
apple_scaled=scaler.fit_transform(apple_tn)
netflix_scaled=scaler.fit_transform(netflix_tn)
google_scaled=scaler.fit_transform(google_tn)
```

```python
meta_scaled
```

# 1. Meta

```python
len(meta_scaled)
```

```python
X  = []
y  = []
for i in range(60,len(meta_scaled)):
    X.append(meta_scaled[i-60:i, 0])
    y.append(meta_scaled[i, 0])

X  = np.array(X )
y  = np.array(y )
print(X.shape)
print(y.shape)
```

```python
split_ratio = 0.7
split_index = int(split_ratio * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
from keras.models import Sequential
from keras.layers import LSTM 
from keras.layers import Dense 
from keras.layers import Dropout
```

```python
split_index
```

```python
reg = Sequential()
reg.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
reg.add(Dropout(0.2))
reg.add(LSTM(units=50))
reg.add(Dropout(0.2))
reg.add(Dense(units=1))
reg.compile(optimizer='adam', loss='mean_squared_error')
reg.summary()
```

```python
his=reg.fit(X_train, y_train, epochs=100, batch_size=64)
```

```python
plt.plot(his.history['loss'])
```

```python
# Make predictions
y_pred_train = reg.predict(X_train)
y_pred_test = reg.predict(X_test)

# Inverse transform to get original scale
y_pred_train = scaler.inverse_transform(y_pred_train)
y_pred_test = scaler.inverse_transform(y_pred_test)
y_train_orig = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test_orig = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
# Plot testing predictions
plt.figure(figsize=(12,8))
plt.plot(y_test_orig, label='Actual Test')
plt.plot(y_pred_test, label='Predicted Test')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title(f' Testing Set Stock Price Prediction ')
plt.legend()
plt.grid()
plt.show()
```

```python
combined_pred = np.concatenate((y_pred_train, y_pred_test))
combined_actual = np.concatenate((y_train_orig, y_test_orig))
plt.figure(figsize=(12,8))
plt.plot(combined_actual, label='Actual')
plt.plot(combined_pred, label='Predicted')
plt.axvline(x=len(y_train_orig), color='gray', linestyle='--', label='Train-Test Split')
plt.xlabel('no. of Days from 01/01/2012')
plt.ylabel('Price')
plt.title(f' Complete Set Stock Price Prediction')
plt.legend()
plt.grid()
plt.show()
```

```python
# Predict next 10 days
last_sequence = X[-1]  # Use the last sequence from the data
predicted_prices = []  # Store the predicted prices
sequence_length = 60
for _ in range(10):
    next_day_prediction = reg.predict(last_sequence.reshape(1, sequence_length, 1))
    predicted_prices.append(next_day_prediction[0][0])
    last_sequence = np.append(last_sequence[1:], next_day_prediction[0][0])
    
# Inverse transform to get original scale
predicted_prices = scaler.inverse_transform(np.array(predicted_prices).reshape(-1, 1))
plt.figure(figsize=(6,6))
# Plot actual prices and predicted prices for the next 10 days
plt.plot(predicted_prices, marker='*')
plt.title(f'Predicted stock price of for next 10 days')
plt.xlabel('Days')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
